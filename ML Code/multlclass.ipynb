```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# GRE based
```

```python
df = pd.read_csv('student.csv')
X_train = df.iloc[ : , 0: 6].values
y_train = df.iloc[:, -1].values
data = [[170, 5, 150, 91, 8.7, 2]]
data = np.array(data)
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(data)
grade = svm_predictions.reshape(1,1)[0]
grade = grade[0]
grade
```

```python
university_df = pd.read_csv('college.csv')
#result = (university_df[university_df.Grade == grade]).sort_values(by = 'Overall Rating', ascending = False)
budget = 50
result = (university_df[(university_df.Grade == grade) & (university_df.Fees_In_Lakhs < budget + 4)]).sort_values(by = 'Overall Rating', ascending = False)
result
```
